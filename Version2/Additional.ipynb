```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    'outlier': {'Recall SVM': 0.75, 'Recall Logistic': 0.75, 'observation': 8},
    'node': {'Recall SVM': 0.8, 'Recall Logistic': 1.0, 'observation': 10},
    'Low Speed Limit, Car vs. Car': {'Recall SVM': 0.714286, 'Recall Logistic': 0.735450, 'observation': 378},
    'Low Speed Limit, Pedestrian vs. Car': {'Recall SVM': 0.9, 'Recall Logistic': 0.9, 'observation': 40},
    'High Speed Limit, Car vs. Car': {'Recall SVM': 0.652174, 'Recall Logistic': 0.782609, 'observation': 46},
    'Origin': {'Recall SVM': 0.669767, 'Recall Logistic': 0.683721, 'observation': 46}
}

df = pd.DataFrame.from_dict(data, orient='index')
df
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

x = range(len(df.index))
bar_width = 0.35

bars1 = ax.bar(x, df['Recall SVM'], width=bar_width, label='Recall SVM')
bars2 = ax.bar([i + bar_width for i in x], df['Recall Logistic'], width=bar_width, label='Recall Logistic')

for i, bar in enumerate(bars1):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height, f'{df["Recall SVM"].iloc[i]:.2f}', ha='center', va='bottom')

for i, bar in enumerate(bars2):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height, f'{df["Recall Logistic"].iloc[i]:.2f}', ha='center', va='bottom')

for i in range(len(df)):
    ax.text(x[i] + bar_width / 2, max(df['Recall SVM'].iloc[i], df['Recall Logistic'].iloc[i]) + 0.02, 
            f'Obs: {df["observation"].iloc[i]}', ha='center', va='bottom', fontsize=10, color='#cf8282')

ax.set_xticks([i + bar_width / 2 for i in x])
ax.set_xticklabels(df.index, rotation=10,ha='center')

ax.set_xlabel('Categories')
ax.set_ylabel('Recall')
ax.set_title('Recall Comparison (SVM vs Logistic)')
ax.legend()

plt.tight_layout()

plt.savefig('../CalculatedData/Plots/recalls.png', bbox_inches='tight')

plt.show()
```
